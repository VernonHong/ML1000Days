```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1.   畫出 cos 圖檔，並儲存

x = np.arange(0,np.pi*3,0.1)
y = np.cos(x)

plt.plot(x,y)
plt.savefig("filename.png",dpi=300,format="png")
```

```python
#2.    給定散點圖顏色
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
t = np.arctan2(Y,X)
plt.scatter(X,Y,c=t,alpha = 0.6,s = 80)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
