```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 
#  "projection" ortho"
#code 
#
plt.figure(figsize=(8,6),edgecolor = 'w')

#
m=Basemap(projection = 'ortho',lat_0 = 0,lon_0 = 0)

#, NASA  
m.bluemarble()

#
plt.show()
```

```python


#Lambert
# width=12000000,height=9000000
#lat_1=45.,lat_2=55,lat_0=50,lon_0=-107
#set resolution = None
# projection='lcc'

m = Basemap(width = 12000000,height=9000000, projection = 'lcc',lat_1 =45,lat_2 = 55,lat_0 = 50,lon_0=-107 ,resolution = None)

#NASA  
m.bluemarble()

#
plt.show()
```

```python

#  "projection" Mill"
# 
#  - llcrnrlat = -90,
#  - llcrnrlon = -180,
#  - urcrnrlat = 90,
#  -urcrnrlon = 180

# resolution

m = Basemap(projection = 'mill',llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180,resolution = 'c') # resolution 

#
m.drawcoastlines()
#2 
m.drawcountries(linewidth = 2)
#
m.drawstates(color='b')
#
m.drawcounties(color='darkred')

plt.title('Basemap Tutorial')
plt.show()
```
