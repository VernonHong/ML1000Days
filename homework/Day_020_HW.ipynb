```python
import numpy as np
import seaborn as sns

#設定圖形樣式 - whitegrid
# use sns.set
sns.set(style = 'whitegrid')
# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
x = np.random.normal(2, 1, 75)
y = 2 + 1.5 * x + np.random.normal(0, 2, 75)


# 畫圖
# sns.residplot
sns.residplot(x,y)
```

```python
sns.displot(y,bins = 20,rug =True,kde = True)
```
